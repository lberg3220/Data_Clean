```python
import pandas as pd
```

```python
file_name = '/content/Bacteroides_Fragilis.csv'
```

```python
df = pd.read_csv(file_name, sep=',')
```

```python
print(df)
```

```python
df = df.copy()
rows_with_nm = df[df['Standard Units'].str.contains('nM')]
rows_with_nm['Standard Value'] = rows_with_nm['Standard Value'] / 1000
rows_with_nm['Standard Units'] = rows_with_nm['Standard Units'].str.replace('nM', 'mM')
df.update(rows_with_nm)
```

```python
print(df)
```

```python
def add_activity_column(df):
  """
  Adds a new column named "Activity" to the dataframe based on the values in the "Standard Units" and "Standard Value" columns.
  """
  df["Activity"] = 3

  # Find the rows where "Standard Units" contains "%"
  percent_rows = df[df["Standard Units"].str.contains("%")]

  mM_rows = df[df["Standard Units"].str.contains("mM")]

  # Update the "Activity" column for those rows
  for i, row in percent_rows.iterrows():
    if row["Standard Value"] >= 90:
      df.loc[i, "Activity"] = 1
    elif row["Standard Value"] <= 0:
      df.loc[i, "Activity"] = 0

  for j, row in mM_rows.iterrows():
    if row["Standard Value"] <= 20:
      df.loc[j, "Activity"] = 1
    elif row["Standard Value"] >= 100:
      df.loc[j, "Activity"] = 0

  return df
```

```python
def add_activity_column_and_remove_other_rows(df):
  """
  Adds a new column named "Activity" to the dataframe based on the values in the "Standard Units" and "Standard Value" columns, and removes rows that don't fall into either category.
  """
  df = add_activity_column(df)

  # Keep only rows where Activity is 0 or 1
  df = df[df["Activity"].isin([0, 1])]

  return df
```

```python
df = df.copy()

# Add the "Activity" column
df = add_activity_column_and_remove_other_rows(df)

# Print the updated dataframe
print(df)
```

```python
def save_as_csv_with_target_name(df):
  """
  Saves the dataframe as a CSV file with the filename as the "Target Name" value.
  """
  target_name = df["Target Name"].iloc[0]
  filename = f"{target_name}.csv"
  df.to_csv(filename, index=False)
```
